```python
"""
Python project

This project receives an license plate number input from the user and classifies it as 
either old style plate number, new style plate number or invalid number

Input: String
Output: String
"""

plate_num = input("Enter license plate number: ")

if len(plate_num) == 6 and plate_num[-3:].isnumeric() and plate_num[:4].isupper():
    print('Number is valid old style license plate')
elif len(plate_num) == 8 and plate_num[:4].isnumeric() and plate_num[5:].isupper():
    print('Number is valid new style license plate')
else:
    print('Invalid number')
```

```python
"""
This programme simulates a spin of a roulette and displays the number that 
was selected and all of the bets that must be payed. 
It only covers the following bets:
- Single number (1 to 36, 0, or 00)
- Red versus Black
- Odd versus Even
- 1 to 18 versus 19 to 36

Input: None
Output: String
"""

import random

green = [0, '00']
red = list(range(1,10, 2)) + list(range(12,19, 2)) + \
list(range(19, 28, 2)) + list(range(30, 37, 2))
black = [n for n in range(1, max(red)) if n not in red]
roulette_wheel = red + black + green

index = random.randint(0, len(roulette_wheel) - 1)
num = roulette_wheel[index]


print(f'The spin resulted in {num}...')
if 0 != num != '00':
    print(f'Pay {num}')
    if num in red:
        print(f'Pay Red')
        
    if num in black:
        print(f'Pay Black')
        
    if num % 2 == 0:
        print('Even')
    else:
        print('Odd')
        
    if 1 <= num <= 18:
        print(f'Pay 1 to 18')
        
    if 19 <= num <= 36:
        print(f'Pay 19 to 36')
else:
    if num == 0:
        print('Pay 0')
    else:
        print('Pay 00')
```
